## Q2

請合併鐵達尼號乘客資料的

1. 訓練資料集 (`titanic_train.csv`)
1. 測試資料集 (`titanic_test.csv`)

提示：

1. 請先確認兩份資料的欄位是否一致
1. 請在測試資料集 (`titanic_test.csv`) 中新增缺失的欄位，並補入 `NaN` 值
1. 請依照 index (row) 方向結合這兩份資料，並且重設 Index Label 為從 `0` 開始的數列

```python
# 讀取訓練資料集、以及測試資料集
df_train = pd.read_csv('titanic_train.csv')
df_test = pd.read_csv('titanic_test.csv')
```

```python
# 排序後比較容易找到兩者的不同
print(df_train.columns.sort_values())
print(df_test.columns.sort_values())
```

```python
# 當欄位很多的時候，可以調用 columns 的 method .difference()，幫助你快速找到兩份資料的差異
df_train.columns.difference(df_test.columns)
```

```python
# 依照 index (row) 方向結合這兩份資料，並且重設 Index Label 為從 0 開始的數列
df_concat = pd.concat([df_train, df_test], axis=0, sort=False)  # 不加上排序的設定會出現警告
df_concat.reset_index(drop=True, inplace=True)
df_concat
```
