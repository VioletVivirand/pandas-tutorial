### Q2. 請合併鐵達尼號乘客資料一（titanic_train.csv）和資料二（titanic_test.csv）

提示：
1. 請先確認兩份資料的欄位是否一致
1. 請在資料二（titanic_test.csv）中新增缺失的欄位，並補值 NaN
1. 請依照 index (row) 方向結合這兩份資料，並且重設 Index Label從 0 開始的數列


```python
# 1. 請先確認兩份資料的欄位是否一致
df_train = pd.read_csv('titanic_train.csv')
df_test = pd.read_csv('titanic_test.csv')

# 排序後比較容易找到兩者的不同
print(df_train.columns.sort_values())
print(df_test.columns.sort_values())
```

```python
# 補充1.：你可以一一比對，但是當欄位很多的時候，可以調用 columns 的 method .difference() 幫助你快速找到
df_train.columns.difference(df_test.columns)
```

```python
# 請依照 index (row) 方向結合這兩份資料，並且重設 Index Label從 0 開始的數列
df = pd.concat([df_train, df_test], axis=0)
df.reset_index(drop=True, inplace=True)
df
```
